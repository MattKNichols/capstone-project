```python
#Calculating a 24-hour rolling average to determine conformance with the NAAQS for PM2.5.
avg_pm25 = pd.rolling_mean(aq_ts.pm25, window=24, min_periods=12, freq=None, center=False)
avg_pm25.tail(25)
```

```python
avg_pm25.describe()
```

```python
#Plotting the rolling 24-hour average PM2.5 values, with the NAAQS shown as the red horizontal line.
fig = plt.figure()
plt.plot(avg_pm25)
plt.plot(('2016-03-15 00:00:00','2017-02-05 20:00:00'),(35,35),'r-')
plt.title('Rolling 24-hr Avg PM2.5 Compared to NAAQS')
plt.tight_layout()
plt.show()
fig.savefig('rolling_pm25.png')
```

```python
avg_o3 = pd.rolling_mean(aq_ts.o3, window=8, min_periods=4, freq=None, center=False)
avg_o3.tail(25)
```

```python
avg_o3.describe()
```

```python
#Plotting the rolling 8-hour average O3 values, with the NAAQS shown as the red horizontal line.
fig = plt.figure()
plt.plot(avg_o3)
plt.plot(('2016-03-15 00:00:00','2017-02-05 20:00:00'),(0.070,0.070),'r-')
plt.title('Rolling 8-hr Avg O3 Compared to NAAQS')
plt.tight_layout()
plt.show()
fig.savefig('rolling_o3.png')
```

```python
#Finding the fourth highest daily maximum 8-hour average O3. It is 0.0679.
top_o3 = avg_o3.sort_values(ascending=False)
top_o3.head(15)
```

```python
#Uploading weather data recorded by the National Weather Service at Columbus Airport.
#Observations are made at 9 minutes til the hour, every hour.
weather = pd.read_excel('columbus_weather.xlsx')
weather = pd.DataFrame(weather)
weather.head(25)
```

```python
#Dropping the first 22 rows to get a clean start for the time series at midnight of March 7, 2016.
weather = weather.drop(weather.index[0:23])
weather.head(25)
```

```python
weather.describe()
```

```python
rng = pd.date_range('03/07/2016', periods=len(weather), freq='H')
```

```python
#Constructing a time series. This also rounds up the hourly observations, e.g. 13:59 becomes 14:00.
weather_ts = weather.set_index(rng)
weather_ts.head()
```

```python
#Replacing '*****' values with 0
weather_ts = weather_ts.replace(to_replace='*****',value=0)
```

```python
weather_ts = weather_ts.replace(to_replace='T*****',value=0)
```

```python
weather_ts = weather_ts.replace(to_replace='***',value='CLR')
```

```python
weather_ts.loc[weather_ts.SPD == 'CLR', 'SPD'] = 0
```

```python
weather_ts.loc[weather_ts.TEMP == '****', 'TEMP'] = 'NaN'
```

```python
weather_ts.loc[weather_ts.DEWP == '****', 'DEWP'] = 'NaN'
```

```python
weather_ts.to_excel('weather_ts.xlsx')
```

```python
#Narrowing down the dataframe to just weather attributes relevant to air quality.
weather_ts = pd.DataFrame(weather_ts, index=weather_ts.index, columns=['SPD','SKC','TEMP','DEWP','PCP01','PCP06','PCP24'])
weather_ts.head(500)
```

```python
final_weather = pd.read_excel('columbus_weather_cleaned2.xlsx')
final_weather = pd.DataFrame(final_weather)
final_weather.head()
```

```python
#Merging the air quality and meteorological data into one dataframe, now that they share a time series index.
merged = aq_ts.join(final_weather)
merged.head()
```

```python
merged.describe()
```

```python
#Plotting the relationship (or lack thereof) between particulate matter and whether it rained in the past HOUR.
#Is there a relationship here? Hard to tell.
plt.scatter(merged.PCP01, merged.pm25)
plt.xlabel('Precipitation in previous hour (inches)')
plt.ylabel('Particulate matter concentration')
plt.title('PM2.5 vs. 1-hr Recent Precipitation')
plt.show()
```

```python
#Plotting the relationship (or lack thereof) between particulate matter and whether it rained in the past DAY.
plt.scatter(merged.PCP24, merged.pm25)
plt.xlabel('Precipitation in previous day (inches)')
plt.ylabel('Particulate matter concentration')
plt.title('PM2.5 vs. 1-day Recent Precipitation')
plt.show()
```

```python
#Plotting the relationship (or lack thereof) between particulate matter and temperature.
#Look at the highest PM value at any given temp.
#Below 50 degrees, people DO NOT leave their houses, nor are forest fires likely.
plt.scatter(merged.TEMP, merged.pm25)
plt.xlabel('Temperature (deg F)')
plt.ylabel('Particulate matter concentration')
plt.title('PM2.5 vs. Temperature')
plt.show()
```

```python
#Parrticulate matter versus wind speed.
#Now that's what I call a signal! Wind blows the smoke away!
plt.scatter(merged.SPD, merged.pm25)
plt.xlabel('Wind speed (mph)')
plt.ylabel('Particulate matter concentration')
plt.title('PM2.5 vs. Wind Speed')
plt.show()
```

```python
#So why is the regression line so boring and flat?
#There are still lots of low PM values at each given wind speed. These overpower the strength of the signal.
sns.regplot(x="SPD", y="pm25", data=merged)
```

```python
#Pure noise. Rain has no clearing effect on ozone.
#And in fact, lightning significantly ADDS to ground level ozone.
plt.scatter(merged.PCP01, merged.o3)
plt.xlabel('Precipitation in previous hour (inches)')
plt.ylabel('Ozone concentration')
plt.title('O3 vs. 1-hr Recent Precipitation')
plt.show()
```

```python
#Plotting the relationship (or lack thereof) between ozone and temperature.
#Again, pretty much noise. Perhaps a peak around 70 degrees, ideal outdoor recreation temps.
plt.scatter(merged.TEMP, merged.o3)
plt.xlabel('Temperature (deg F)')
plt.ylabel('Ozone concentration')
plt.title('O3 vs. Temperature')
plt.show()
```

```python
#Since wind did such a good job dissipating PM, how about ozone? NOPE.
plt.scatter(merged.SPD, merged.o3)
plt.xlabel('Wind speed (mph)')
plt.ylabel('O3 concentration')
plt.title('O3 vs. Wind Speed')
plt.show()
```
