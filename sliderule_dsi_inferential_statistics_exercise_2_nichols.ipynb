# Examining Racial Discrimination in the US Job Market

### Background
Racial discrimination continues to be pervasive in cultures throughout the world. Researchers examined the level of racial discrimination in the United States labor market by randomly assigning identical résumés to black-sounding or white-sounding names and observing the impact on requests for interviews from employers.

### Data
In the dataset provided, each row represents a resume. The 'race' column has two values, 'b' and 'w', indicating black-sounding and white-sounding. The column 'call' has two values, 1 and 0, indicating whether the resume received a call from employers or not.

Note that the 'b' and 'w' values in race are assigned randomly to the resumes when presented to the employer.

1. A two-sample test is appropriate for this problem. The central limit theorem (CLT) applies, because we are dealing with two Bernoulli Distributions. As the sample sizes for both populations (black-sounding names and non-black sounding names) increase, the sample means should approach the population means.

2. The null hypothesis is that there is no difference between the rate of callbacks for black-sounding names and non-black sounding names. The alternative hypothesis is there is a difference between the rate of callbacks for the two groups. To be clear, under the alternative hypothesis, the callback rate for blacks could be higher or lower than for non-blacks.

3. 9.7% of white applicants received a callback, versus 6.4% of black applicants. Thus, the difference between the sample means is 3.3 percentage points, or 0.033, with a standard deviation for the distribution of the sample differences of 0.00778 and a margin of error of 0.015. We can reasonably confident that, should this experiment be repeated an infinite number of times, 95% of the differences between the callback rate for blacks and whites would fall between 0.018 and 0.048, or between 1.8% and 4.8%. For the null hypothesis to be true (no difference between blacks and whites) would require a value 2.2 standard deviations from the mean. This corresponds to a p-value of 0.9861. 

4. The observed difference between black and white applicants is statistically significant at the 0.98 significance level. In other words, should the experiment be repeated an infinite number of times, a difference this large or larger would be expected to occur by chance in less than 2% of cases. 

5. Race, or perceived race via the proxy of name, is proven to be an important factor in job application callbacks. Applicants with black-sounding names received 3.3 percentage points fewer callbacks (6.4% versus 9.7%) than applicants without black-sounding names. As a proportion, this means non-blacks received 49.7% more callbacks.
